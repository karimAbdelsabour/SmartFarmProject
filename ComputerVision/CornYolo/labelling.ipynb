```python
import os

# Specify the directory containing your images
image_folder = r'C:\Users\ziads\Downloads\New folder (8)\SmartFarmProject\ComputerVision\CornYolo\Corn_(maize)___Northern_Leaf_Blight'  # Use raw string

# Initialize a flag to track if any files were found
files_found = False

# Loop through each file in the directory
for filename in os.listdir(image_folder):
    if filename.endswith('.JPG') or filename.endswith('.jpg') or filename.endswith('.png'):
        files_found = True  # Mark that we've found at least one file
        print(f'Processing: {filename}')  # Print the filename being processed
        
        # Create the corresponding .txt filename
        txt_filename = os.path.splitext(filename)[0] + '.txt'
        txt_file_path = os.path.join(image_folder, txt_filename)

        # Create and write the annotation to the .txt file
        with open(txt_file_path, 'w') as f:
            f.write('3 0.500000 0.500000 1.000000 1.000000\n')

# Check if any files were found
if not files_found:
    raise FileNotFoundError("No image files found in the specified directory.")

print("YOLO annotations created successfully.")
```
